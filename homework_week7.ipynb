```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
from sklearn import linear_model
url='/content/homework/passenger1.csv'
data=pd.read_csv(url)
data.hist()
plt.show()
```

```python
sns.pairplot(data)
```

```python
correlation=data.corr()
print(correlation)
```

```python
sns.heatmap(correlation,annot=True)
```

```python
data.plot(kind='scatter',x='Miles(X)',y='Dollars (Y)',figsize=(9,9),color='purple')
```

```python
from sklearn import linear_model
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
from sklearn import linear_model
data.plot(kind='scatter',x='Miles(X)',y='Dollars (Y)',figsize=(9,9),color='purple')
model=linear_model.LinearRegression()
model.fit(X = pd.DataFrame(data['Miles(X)']),y = data['Dollars (Y)'])
print('Giá trị intercept:',model.intercept_)
print('Giá trị coeffient:',model.coef_)
print('Độ chính xác:',model.score(X=pd.DataFrame(data['Miles(X)']),y=data['Dollars (Y)']))
```

```python
data.plot(kind='scatter',x='Miles(X)',y='Dollars (Y)',figsize=(9,9),color='black',xlim=(0,7))
y_pred=model.predict(X=pd.DataFrame(data['Miles(X)']))
plt.plot(data['Miles(X)'],y_pred,color='blue')
```

```python
model=linear_model.LinearRegression()
X=pd.DataFrame([data['Miles(X)'],data['Dollars (Y)']**2]).T
model.fit(X,y=data['Dollars (Y)'])
print('Intercept là:',model.intercept_)
print('Coeficient là:',model.coef_)
print('Độ chính xác là:',model.score(X,y=data['Dollars (Y)']))
range = np.arange(1.5,7,0.1)
X_range=pd.DataFrame([range,range**2]).T
y_pred=model.predict(X=X_range)
data.plot(kind='scatter',x='Miles(X)',y='Dollars (Y)',figsize=(9,9),color='black',xlim=(0,7))
plt.plot(range,y_pred,color='blue')
```

```python
model=linear_model.LinearRegression()
X=pd.DataFrame([data['Miles(X)'],data['Miles(X)']**2,data['Miles(X)']**3,data['Miles(X)']**4,data['Miles(X)']**5,data['Miles(X)']**6,data['Miles(X)']**7,data['Miles(X)']**8,data['Miles(X)']**9,data['Miles(X)']**10]).T
model.fit(X,y=data['Dollars (Y)'])
print('Intercept là:',model.intercept_)
print('Coeficient là:',model.coef_)
print('Độ chính xác là:',model.score(X,y=data['Dollars (Y)']))
range = np.arange(1.5,9,0.1)
X_range=pd.DataFrame([range,range**2,range**3,range**4,range**5,range**6,range**7,range**8,range**9,range**10]).T
y_pred = model.predict(X = X_range)
data.plot(kind='scatter',x='Miles(X)',y='Dollars (Y)',figsize=(9,9),color='black',xlim=(0,7))
plt.plot(range,y_pred,color='blue')
```
